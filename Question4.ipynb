```python
# Importing the Libraries
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras import layers
import matplotlib.pyplot as plt
from keras.datasets import mnist

class Question4:
  def __init__(self, data):
    (self.xTrain, self.yTrain), (self.xTest,self.yTest) = data.load_data()
  
  def createModel(self):
    self.model=keras.Sequential(name='Digit_pattern_recognition')
    self.model.add(layers.Flatten(input_shape=(28,28)))
    self.model.add(layers.Dense(256, activation = "relu", name = "HiddenLayer1"))
    self.model.add(layers.Dense(64, activation = "relu", name = "HiddenLayer2"))
    self.model.add(layers.Dense(10, activation="softmax",name= "OutputLayer"))
    
  def RunModel(self):
    self.model.compile(loss = keras.losses.SparseCategoricalCrossentropy(from_logits=True), optimizer = "adam", metrics = ["accuracy"])
    self.fitOutput = self.model.fit(self.xTrain,self.yTrain)
    
  def evaluate(self):
    self.output = self.model.fit(self.xTest, self.yTest)
  
  def showGraphs(self):
    plt.figure(figsize=(10,12))
    for i in range(25):
      plt.subplot(5,5,i+1)
      plt.xticks([])
      plt.yticks([])
      plt.grid(False)
      plt.imshow(self.xTrain[i],cmap=plt.cm.binary)
      plt.xlabel(self.yTrain[i])
    plt.show()

  

if __name__ == "__main__":
  object_ = Question4(mnist)

  # Showing the MNIST data
  object_.showGraphs()

  # Creating the model
  object_.createModel()

  # fit and run the model
  object_.RunModel()

  # Evaluate the model
  object_.evaluate()




```
